# Question **18**

```python
print(df.isnull().sum())  # Displays count of missing values in each column
```

```python
df["education"].fillna(0, inplace=True)
df["marital"].fillna(0, inplace=True)
df["y"].fillna(0, inplace=True)
df.drop(columns=["marital_status"], inplace=True)
```

```python
df["y"] = df["y"].map({"yes": 1, "no": 0})
df_encoded = pd.get_dummies(df, drop_first=True)
corr_matrix = df_encoded.corr()
print(corr_matrix["y"].sort_values(ascending=False))
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Banking Dataset")
plt.show()
```

```python
# Sort correlation values
corr_with_y = corr_matrix["y"].sort_values(ascending=False)

# Select top 10 positive and negative correlations (excluding 'y' itself)
top_features = pd.concat([corr_with_y.head(11), corr_with_y.tail(10)])  # 10 highest & 10 lowest

# Plot the bar chart
plt.figure(figsize=(12, 6))
top_features.drop("y").plot(kind="barh", color=['green' if x > 0 else 'red' for x in top_features.drop("y")])
plt.xlabel("Correlation with Subscription (y)")
plt.ylabel("Features")
plt.title("Top Features Correlated with Term Deposit Subscription", fontsize=14)
plt.axvline(x=0, color="black", linestyle="--")  # Reference line at 0
plt.show()
```
